```python
df3 = pd.merge(df1, df2, on="date")
df3
```

```python
import c3aidatalake
```

```python
# Fetch participants who are located in California and who have a relatively strong intent to wear a mask in public because of COVID-19
survey = c3aidatalake.fetch(
    "surveydata",
    {
        "spec": {
            "filter": "location == '{}_UnitedStates'".format(state.replace(" ", ""))
        }
    },
    get_all = True
)

survey.columns
```

```python
survey.dropna(inplace=True)
survey = survey[["coronavirusIntent_Mask", "coronavirusIntent_SixFeet","startTime"]].mean()
survey
```

```python
df4 = pd.concat([df3, 
                 pd.Series(survey["coronavirusIntent_Mask"], 
                           index=df3.index, name='mask_intent'),
                 pd.Series(survey["coronavirusIntent_SixFeet"], 
                           index=df3.index, name='social_distance_intent')], axis=1)
df4
```

```python
mask_mandate = pd.read_csv("datasets\mask_mandates.txt", sep=",")
# print(mask_mandate.columns)
mask_mandate["state_wide_effect"] = pd.to_datetime(mask_mandate["state_wide_effect"].astype(str), format='%m/%d/%Y')
# pd.to_datetime(mask_mandate["state_wide_effect"]).dt.strftime('%m/%d/%Y')
m_date = mask_mandate[mask_mandate["State"]==state]["state_wide_effect"]
m_date = m_date.values[0]
# mask_mandate
```

```python
df5 = pd.concat([df4, 
                 pd.Series(0, index=df4.index, name='mask_mandate')], axis=1)
df5['date'] = pd.to_datetime(df5["date"].astype(str))
# print(df5.dtypes)
df5.loc[df5['date']>=m_date, 'mask_mandate'] = 1
df5[['date', 'mask_mandate']]
```
